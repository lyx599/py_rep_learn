# 数据可视化

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# 因子变量绘图

## boxplot箱线图

当对所有数据画箱线图时，只对数值起作用

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('whitegrid')
ax = sns.boxplot(x=tips['total_bill'])
```

```python
#竖放的箱线图
ax = sns.boxplot(y=tips['total_bill'])
```

```python
# 分组绘制箱线图，分组因子是day，在x轴不同位置绘制
ax = sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# 分组箱线图，分子因子是smoker，不同的因子用不同颜色区分
# 相当于分组之后又分组
ax = sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips)
```

```python
#palette 改变配色
ax = sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,palette='Set3')
```

```python
#改变线宽，linewidth参数
ax = sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,linewidth=2.5)
```

```python
# 改变x轴顺序，order参数
ax = sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,order=['Sun','Sat','Fri','Thur'])
```

```python
# 对dataframe的每个变量都绘制一个箱线图，水平放置
iris = sns.load_dataset("iris")
ax = sns.boxplot(data=iris, orient="h", palette="Set2")
```

```python
#垂直放置
ax = sns.boxplot(data=iris, orient="v", palette="Set2")
```

```python
#箱线图+有分布趋势的散点图–>的组合图
# 图形组合也就是两条绘图语句一起运行就可以了，相当于图形覆盖了
ax = sns.boxplot(x="day", y="total_bill", data=tips)
ax = sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
```

## violinplot小提琴图

小提琴图其实是`箱线图`与`核密度图`的结合，箱线图展示了分位数的位置，小提琴图则展示了任意位置的密度，通过小提琴图可以知道哪些位置的密度较高。在图中，白点是中位数，黑色盒型的范围是下四分位点到上四分位点，细黑线表示须。外部形状即为核密度估计（在概率论中用来估计未知的密度函数，属于非参数检验方法之一）

```python
# 绘制小提琴图
ax = sns.violinplot(x=tips["total_bill"])
```

```python
# 分组的小提琴图，同上面的箱线图一样通过X轴分组
ax = sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
# 通过hue分组的小提琴图，相当于分组之后又分组
ax = sns.violinplot(x="day", y="total_bill",hue='smoker', data=tips)
```

```python
# 分组组合的小提琴图，其实就是hue分组后，各取一半组成一个小提琴图
ax = sns.violinplot(x="day", y="total_bill", hue="smoker",data=tips, split=True)
```

## stripplot散点图

需要注意的是，seaborn中有两个散点图，一个是普通的散点图，另一个是可以看出分布密度的散点图。  
下面把它们画在一起就明白了。

```python
# 普通的散点图
ax1 = sns.stripplot(x=tips["total_bill"])
```

```python
# 带分布密度的散点图
ax2 = sns.swarmplot(x=tips["total_bill"])
```

```python
# 分组的散点图
ax = sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
# 添加抖动项的散点图，jitter可以是0.1,0.2...这样的小数，表示抖动的程度大小,默认是抖动的，Flase表示不抖动
ax = sns.stripplot(x="day", y="total_bill", data=tips,jitter=False)
```

```python
#横着放，x-y顺序换一下就好了
ax = sns.stripplot(x="total_bill", y="day", data=tips)
```

```python
# 分组
ax = sns.stripplot(x="day", y="total_bill", hue="smoker", data=tips)
```

```python
#分组后分开绘制 dodge参数
ax = sns.stripplot(x="day", y="total_bill", hue="smoker",data=tips, dodge=True)
```

```python
# 散点图+小提起图
ax = sns.violinplot(x="day", y="total_bill", data=tips,inner=None, color=".9")
ax = sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

# 回归图

regplot()和lmplot()它们紧密相关，而且共享大多数的核心功能。具体有什么差别还不知道
  
  
最主要的区别在于regplot()中的x, y参数接受多种数据类型，包括numpy数组、pandas序列（Series），或者将pandas DataFrame传递给data参数。作为对比，lmplot()的data参数是不能为空的，同时x和y参数必须以字符串形式指定。这种数据格式（这里是指regplot()支持而lmplot()不支持的类似一维数组的数据格式）被称为“long-form data”或“tidy data”。

## 回归图lmplot

```python
ax = sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# 分组的线性回归图，通过hue参数控制
ax = sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips)
```

```python
# 分组绘图，不同的组用不同的形状标记
ax = sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips,markers=['o','x'])
```

```python
# 不仅分组，还分开不同的子图绘制，用col参数控制
ax = sns.lmplot(x='total_bill',y='tip',col='smoker',data=tips)
```

```python
# col+hue 双分组参数，既分组，又分子图绘制
ax = sns.lmplot(x='total_bill',y='tip',hue='day',col='day',data=tips)
```

```python
#分组绘制，控制height尺寸
ax = sns.lmplot(x="total_bill", y="tip", col="day", hue="day",
                data=tips, col_wrap=2, height=5)
```

```python
# 既然col可以控制分组子图的，那么row也是可以控制分组子图的
ax = sns.lmplot(x="total_bill", y="tip", row="sex", 
                col="time", data=tips, height=3)
```

```python
ax = sns.lmplot(x="total_bill", y="tip", row="day",
                data=tips, height=5)
```

col控制列，col有col_warp控制一行有多少列，row控制行

```python
ax = sns.lmplot(x="total_bill", y="tip", col="day", row="smoker",
                data=tips, height=3)
```

## 回归图regplot

```python
ax = sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
# 控制回归的置信度，你会看到拟合直线的外面的面积的有变化的
ax = sns.regplot(x="total_bill", y="tip", data=tips,ci=68)
```

```python
# 上面的都是拟合一次曲线，拟合二次曲线通过order=2设置，
# 拟合一次曲线相当于 order=1
ans = sns.load_dataset("anscombe")
ax = sns.regplot(x="x", y="y", data=ans.loc[ans.dataset == "II"],
                scatter_kws={"s": 80},order=2, ci=None, truncate=True)
```

## 核密度图kdeplot

```python
np.random.seed(10)
mean, cov = [0, 2], [(1, .5), (.5, 1)]
x, y = np.random.multivariate_normal(mean, cov, size=50).T
ax = sns.kdeplot(x)
```

```python
# shade参数决定是否填充曲线下面积
ax = sns.kdeplot(x, shade=True, color="r")
```

```python
# 双变量密度图，相当于等高线图了
# shade 参数改用颜色深浅表示密度的大小，不过不用，就真的是等高线了
ax = sns.kdeplot(x, y, shade=True)
```

```python
ax = sns.kdeplot(x, y)
```

```python
# 分组绘制双变量的核密度图
# 相当于绘制两个核密度图，通过图可以看到密度中心
# 类似于挖掘算法中聚类中心绘图
setosa = iris.loc[iris.species == "setosa"]  # 组1
virginica = iris.loc[iris.species == "virginica"]  # 组2

ax = sns.kdeplot(setosa.sepal_width, setosa.sepal_length, 
                    cmap="Reds", shade=True, shade_lowest=False)

ax = sns.kdeplot(virginica.sepal_width, virginica.sepal_length, 
                    cmap="Blues", shade=True, shade_lowest=False)
```

# 双坐标轴图


适用于观察两个变量在同一幅图上的变化趋势，而两个变量数值差距大，需要双坐标轴

```python
sale=pd.Series(np.random.random(10)*100).map(int)
tmperature=pd.Series(np.random.random(10)*10).map(int)

ax=plt.subplot(111)
sale.plot(ax=ax,color='b')
ax.set_xlabel('time')
ax.set_ylabel('sale')
# 重点来了，twinx 或者 twiny 函数
ax2 = ax.twinx()
tmperature.plot(ax=ax2,color='r')
ax2.set_ylabel('tmperature')
plt.title('double series figure')
```
