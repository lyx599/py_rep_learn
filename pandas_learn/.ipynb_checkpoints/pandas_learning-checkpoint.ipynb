# 4.pandas设置值

```python
# 根据位置设置loc和iloc
df.iloc[2,2] = 111
df.loc['20190316','B'] = 2222
df
```

```python
# 根据条件设置
# 更改B中的数，而更改的位置取决于4的位置，并设相应位置的数为0
df.B[df.A>4] = 0
df
```

```python
df.B.loc[df.A>4] = 0
df
```

```python
# 按行或列设置
# 列批处理，F列全改为NaN
df['F'] = np.nan
df
```

```python
#添加Series序列(长度必须对齐)
df['E'] = pd.Series([1,2,3,4,5,6],index=pd.date_range('20190316',periods=6))
df
```

# 5.处理丢失数据

```python
dates = pd.date_range('20180820',periods=6)
df = pd.DataFrame(np.arange(24).reshape((6,4)),index=dates,columns=['A','B','C','D']) 
df
```

```python
df.iloc[0,1] = np.nan
df.iloc[1,2] = np.nan
df
```

```python
# 删除掉含有NaN的行或列
df.dropna() #默认删除掉含有NaN的行
```

```python
df.dropna(
    axis=1, # 0对行进行操作;1对列进行操作
    how='any' # 'any':只要存在NaN就drop掉;默认是any；'all':必须全部是NaN才drop
)
```

```python
# 替换NaN值为0或者其他
df.fillna(value=0)
```

```python
# 是否有缺失数据NaN
# 是否为空
df.isnull()
```

```python
# 是否为NaN
df.isna()
```

```python
# 检测某列是否有缺失数据NaN
df.isnull().any()
```

```python
# 检测数据中是否存在NaN,如果存在就返回True
np.any(df.isnull()==True)
```

# 7.pandas合并操作

```python
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1, columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*2, columns=['a','b','c','d'])
df1
```

```python
df2
```

```python
df3
```

```python
# concat纵向合并
res = pd.concat([df1,df2,df3],axis=0)
res
```

```python
# 上述合并过程中，index重复，下面给出重置index方法
# 只需要将index_ignore设定为True即可
res = pd.concat([df1,df2,df3],axis=0,ignore_index=True)

# 打印结果
res
```
